```python
######################################################################
# Semi-Supervised Classification of Graph Nodes using Exponential Decay
# L42: Assessment 2
# Jan Ondras (jo356), Trinity College
######################################################################
#############################################################################################################
# Plot graphs of the citation networks used
# maybe redo with 15x7
#############################################################################################################
import numpy as np
import networkx as nx
from matplotlib import pyplot as plt
import time
from gcn.utils import *

dataset_type = 'cora'
# Load data
adj, _, _, _, _, _, _, _ = load_data(dataset_type)
# Get graph of the network
G = nx.from_scipy_sparse_matrix(adj)
plt.figure(figsize=(15,15))
pos=nx.spring_layout(G)
nx.draw(G, pos, node_size=5)
plt.show()
```
